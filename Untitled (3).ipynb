```python
idade = 20

def calc_faixa_etaria(idade):
    
    if np.isnan(idade):
        return "desconnhecida"
    elif idade <= 15:
        return "0-15"
    elif idade <= 30:
        return "15-30"
    elif idade <= 45:
        return "30-45"
    else:
        return "45+"
```

```python
calc_faixa_etaria(6)
```

```python
df["age"].isnull().sum()
```

```python
df["age"]
```

```python
# quantos passageiros há em cada faixa etária? (0-15, 15-30, 30-45, 45+) (fechado superior, aberto inferior)

df["age_group"] = df["age"].apply(calc_faixa_etaria)
```
